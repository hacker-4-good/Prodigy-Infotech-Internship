```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('world_bank.csv')
data.head()
```

```python
data = data.drop(columns=['Indicator Name','Indicator Code','2022', 'Unnamed: 67'])
data.head()
```

```python
data.isnull().sum()
```

```python
for i in range(1960, 2022):
    data[str(i)] = data[str(i)].fillna(np.mean(data[str(i)]))
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data.describe()
```

```python
plot_data = data.drop(columns=['Country Name', 'Country Code'])
```

```python
work = np.transpose(data)
plt.figure(figsize=(30,6))
plt.xlabel('Years', fontsize=18)
plt.title(work[0]['Country Name'], fontsize=25)
sns.barplot(x=work.index[2:], y=work[2:][0])
plt.show()
```
